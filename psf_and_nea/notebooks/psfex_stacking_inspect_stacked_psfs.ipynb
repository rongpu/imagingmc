```python
from __future__ import division, print_function
import sys, os, glob, time, warnings, gc
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, vstack, hstack, join
import fitsio
# from astropy.io import fits

from matplotlib.colors import LogNorm
```

```python
params = {'legend.fontsize': 'x-large',
          'axes.labelsize': 'x-large',
          'axes.titlesize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large',
          'figure.facecolor': 'w'}
plt.rcParams.update(params)
```

```python
cat = Table(fitsio.read('/global/cfs/cdirs/desi/users/rongpu/imaging_mc/nea/stacked_psfex_g_1024.fits'))

for index in range(1, len(cat), 10):
    if cat['fwhm_mean'][index]!=0:
        print(cat['fwhm_bin'][index])
        fig, ax = plt.subplots(1, 2, figsize=(10, 5))
        ax[0].imshow(cat['psf_mask'][index], cmap='gray')
        im = ax[1].imshow(cat['psf_mask'][index]+1e-7, cmap='gray', norm=LogNorm())
        # plt.savefig('plots/{}.png'.format(index))
        plt.show()
```
